```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 

df = pd.read_csv("ex1data1.txt",delimiter=",")
df.head(3)
m = 97
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['profit'],df['population'],'rx',markersize=10)
plt.ylabel('Profit in $10,000s')
plt.xlabel('Population of City in 10,000s')
plt.xlim([5,25])
plt.ylim([-5,25])
plt.grid(True)
```

```python
df['theta0'] = 1
df.head(3)
```

```python
X = np.array(df[['theta0','profit']])
y = np.array(df[['population']])
```

#### _Cost Function_ _J($\theta$)_ 

```python

def hyp(theta,X):
    return np.dot(X,theta)

def cost_function(theta,X,y):
    hx = hyp(theta,X) - y
    return float((np.dot(hx.T,hx))/(2*97))

intial_theta = np.zeros([2,1])

cost_function(intial_theta,X,y)
```

```python
hyp(theta,X).shape 
X.shape
np.dot(X.T , hyp(theta,X))
theta = intial_theta 
```

### Actual Gradient Descent 

```python
alpha = 0.01
(alpha/m)*np.dot(X.T , hyp(theta,X) - y)   
```

```python
theta = intial_theta  
cost_func_graph = []
epochs = 1000
for i in range(epochs):
    theta = theta - (alpha/m)*(np.dot(X.T , hyp(theta,X) - y))
    cost_func_graph.append(cost_function(theta,X,y))
```

```python
plt.figure(figsize=(10,6))
plt.plot([i for i in range(epochs)],cost_func_graph)
plt.xlabel("Epochs")
plt.ylabel("J ( theta )   /   Cost Function")
plt.grid(True)
```
